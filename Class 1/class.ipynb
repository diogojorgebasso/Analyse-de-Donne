# Class 1
Un jardinier aimerait savoir si les fleurs qu'il a plantées sur son terrain suivent bien les spécificités de la notice qu'il a reçue lorsqu'il a commandé ses graines sur internet. Il était indiqué sur la notice que chaque gousse à maturité doit mesurer 15cm de long.

Comment peut-il s'assurer que les gousses qu'il a dans son jardin suivent bien cette spécificité?

Soit **X**, la variable "longueur". On suppose que X suit une loi Normale de moyenne µ et de variance σ². Nous allons faire un test de Student sur la moyenne µ pour répondre à la question posée.

Nous voulons tester :
- H0 : µ = 15
- H1 : µ ≠ 15


```r
# Simulons les données pour les longueurs des gousses (par exemple) :
set.seed(123) # Pour rendre les résultats reproductibles
glycine <- data.frame(taille = rnorm(54, mean = 15, sd = 1)) # Simulation de 54 gousses avec une moyenne de 15 cm

# Effectuons le test de Student avec la commande t.test() :
resultat_test <- t.test(glycine$taille, mu = 15)
resultat_test
```

# Exemple 2 :
Chez un groupe de 10 sujets, les effets d'un traitement destiné à diminuer la pression artérielle ont été expérimentés. Les résultats (valeur de la tension artérielle systolique en cmHg) ont été relevés sur les 10 sujets avant et après traitement.


```r
# Entrons les observations dans deux vecteurs nommés respectivement 'avant' et 'apres' :
avant <- c(15,18,17,20,21,18,17,15,19,16)
apres <- c(12,16,17,18,17,15,18,14,16,18)

# Calculons la différence entre les deux
difference <- apres - avant
difference
```

```r
# Effectuons le test de Student sur la différence
test_difference <- t.test(difference)
test_difference
```

## Conclusion :
Le résultat du test statistique nous permet de savoir si le traitement a eu un effet significatif sur la pression artérielle avec un seuil de 5%.
